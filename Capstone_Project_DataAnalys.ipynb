```python
# 1. The percentage of road accidents during all the years.


# 'States/Uts' as index
df3.set_index('States/Uts', inplace=True)

# total accidents of each year
yearly_totals = df3.sum()

# percentage of each state
percentages = df3.div(yearly_totals) * 100

# average percentage of all years of each state
avg_percentages = percentages.mean(axis=1).sort_values(ascending=False)

# Ploting of the results
plt.figure(figsize=(12, 8))
avg_percentages.plot(kind='bar')
plt.title('Average Percentage of Road Accidents by State/UT (2003-2016)')
plt.xlabel('States/UTs')
plt.ylabel('Average Percentage')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# top 10 states with the highest average percentage of accident
print("Top 10 States/UTs with Highest Average Percentage of Road Accidents (2003-2016):")
print(avg_percentages.head(10))

# average percentage of each year
yearly_avg_percentages = percentages.mean()

# Ploting the yearwise trend
plt.figure(figsize=(12, 6))
yearly_avg_percentages.plot(kind='line', marker='o')
plt.title('Yearly Average Percentage of Road Accidents (2003-2016)')
plt.xlabel('Year')
plt.ylabel('Average Percentage')
plt.grid(True)
plt.tight_layout()
plt.show()

print("\
Year wise Average Percentages:")
print(yearly_avg_percentages)
```

```python
# 2. Mean Accidents per 1L population for each year

# mean accidents per 1L population of each year
mean_accidents_per_1L = {
    2013: df['Total Number of Accidents Per Lakh Population - 2013'].mean(),
    2014: df['Total Number of Accidents Per Lakh Population - 2014'].mean(),
    2015: df['Total Number of Accidents Per Lakh Population - 2015'].mean(),
    2016: df['Total Number of Accidents Per Lakh Population - 2016'].mean()
}

# Creating a DataFrame from the dictionary for understanding 
mean_accidents_df = pd.DataFrame.from_dict(mean_accidents_per_1L, orient='index', columns=['Mean Accidents per 1L population'])

# Display the results of dataframe
print("Mean Accidents per 1 Lakh Population for each year:")
print(mean_accidents_df)

# Ploting the results
plt.figure(figsize=(10, 6))
mean_accidents_df.plot(kind='bar', ylabel='Mean Accidents per 1 Lakh Population', title='Mean Road Accidents per 1 Lakh Population (2013-2016)')
plt.xlabel('Year')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# the percentage change from 2013 to 2016
percent_change = ((mean_accidents_df.iloc[-1, 0] - mean_accidents_df.iloc[0, 0]) / mean_accidents_df.iloc[0, 0]) * 100

print(f"\
Percentage change from 2013 to 2016: {percent_change:.2f}%")

print("\
Summary Statistics:")
print(mean_accidents_df.describe())
```

```python
#6. The number of accidents happening per state from the year 2003 to 2016.

# Melt the dataframe to convert years to a single column
df_melted = df3.melt(id_vars=['States/Uts'], 
                            var_name='Year', 
                            value_name='Accidents')

# line plot for top 10 states by total accidents
top_10_states = df3.set_index('States/Uts').sum(axis=1).nlargest(10).index

plt.figure(figsize=(15, 8))
for state in top_10_states:
    state_data = df_melted[df_melted['States/Uts'] == state]
    plt.plot(state_data['Year'], state_data['Accidents'], marker='o', label=state)

plt.title('Number of Accidents by State (2003-2016) - Top 10 States')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# total accidents of each state over the entire period
total_accidents = df3.set_index('States/Uts').sum(axis=1).sort_values(ascending=False)
print("\
Total accidents by state (2003-2016):")
print(total_accidents)
```

```python
# 7. Number of ACCIDENTS for 1,2,3,4 LANE per 1L population of resp. state.


# columns of accidents per 1L people
accident_columns = ['Single Lane - Accident - 2014 per 1L people',
                    'Two Lanes - Accident - 2014 per 1L people',
                    '3 Lanes or more w.o Median - Accident - 2014 per 1L people',
                    '4 Lanes with Median - Accident - 2014 per 1L people']

# average accidents per road type
avg_accidents = df4[accident_columns].mean().sort_values(ascending=False)

# bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x=avg_accidents.index, y=avg_accidents.values)
plt.title('Average Accidents per 100,000 People by Road Type (2014)')
plt.xticks(rotation=45)
plt.ylabel('Accidents per 100,000 People')
plt.tight_layout()
plt.show()

print("\
Average accidents per 100,000 people by road type:")
for road_type, value in avg_accidents.items():
    print(road_type.split(' - ')[0] + ": " + str(round(value, 2)))
```

```python
# 8. Number of people INJURED for 1,2,3,4 type of lane per 1L population of resp. State.


# Select relevant columns for injuries per 1L people
injury_columns = ['Single Lane - Injured - 2014 per 1L people',
                 'Two Lanes - Injured - 2014 per 1L people',
                 '3 Lanes or more w.o Median - Injured - 2014 per 1L people',
                 '4 Lanes with Median - Injured - 2014 per 1L people']

# average injuries per road type
avg_injuries = df4[injury_columns].mean().sort_values(ascending=False)

# bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x=avg_injuries.index, y=avg_injuries.values)
plt.title('Average Injuries per 100,000 People by Road Type (2014)')
plt.xticks(rotation=40)
plt.ylabel('Injuries per 100,000 People')
plt.tight_layout()
plt.show()

print("\
Average injuries per 100,000 people by road type:")
for road_type, value in avg_injuries.items():
    print(road_type.split(' - ')[0] + ": " + str(round(value, 2)))
```

```python
# 9. Number of people KILLED for 1,2,3,4 LANES per 1L population of resp. States.

# Select relevant columns for fatalities
columns_of_interest_fatalities = [
    'State/UT',
    'Single Lane - Killed - 2014 per 1L people',
    'Two Lanes - Killed - 2014 per 1L people',
    '3 Lanes or more w.o Median - Killed - 2014 per 1L people',
    '4 Lanes with Median - Killed - 2014 per 1L people'
]

df_fatalities = df4[columns_of_interest_fatalities].copy()

# Remove the 'Total' row
df_fatalities = df_fatalities[df_fatalities['State/UT'] != 'Total ']

# Sort states by total fatalities per 1L people across all lane types
df_fatalities['Total'] = df_fatalities.iloc[:, 1:].sum(axis=1)
df_fatalities_sorted = df_fatalities.sort_values('Total', ascending=False).head(15)

# Create a stacked bar plot for fatalities
fig, ax = plt.subplots(figsize=(15, 10))
df_fatalities_sorted.plot(x='State/UT', y=columns_of_interest_fatalities[1:], kind='bar', stacked=True, ax=ax)

plt.title('Number of People Killed per 1 Lakh Population by Lane Type (Top 15 States/UTs)', fontsize=16)
plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Fatalities per 1 Lakh Population', fontsize=12)
plt.legend(title='Lane Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('fatalities_by_lane_type.png')
plt.close()

print("Graph has been saved as 'fatalities_by_lane_type.png'")

# Display the data for the top 15 states
print("\
Top 15 States/UTs by Total Fatalities per 1 Lakh Population:")
print(df_fatalities_sorted.to_string(index=False))
```

```python
# 10. Number of Accidents, people KILLED, INJURED on SINGLE LANE per 1L population.


# relevant columns for single-lane roads
columns_of_interest = [
    'State/UT',
    'Single Lane - Accident - 2014 per 1L people',
    'Single Lane - Killed - 2014 per 1L people',
    'Single Lane - Injured - 2014 per 1L people'
]

df_single_lane = df4[columns_of_interest].copy()

# Remove the 'Total' row
df_single_lane = df_single_lane[df_single_lane['State/UT'] != 'Total ']

# states by total incidents (accidents + killed + injured) on single-lane roads
df_single_lane['Total'] = df_single_lane.iloc[:, 1:].sum(axis=1)
df_single_lane_sorted = df_single_lane.sort_values('Total', ascending=False).head(15)

# Create a grouped bar plot
fig, ax = plt.subplots(figsize=(15, 10))
x = range(len(df_single_lane_sorted))
width = 0.25

plt.bar(x, df_single_lane_sorted['Single Lane - Accident - 2014 per 1L people'], width, label='Accidents', align='center')
plt.bar([i + width for i in x], df_single_lane_sorted['Single Lane - Killed - 2014 per 1L people'], width, label='Killed', align='center')
plt.bar([i + 2 * width for i in x], df_single_lane_sorted['Single Lane - Injured - 2014 per 1L people'], width, label='Injured', align='center')

plt.xlabel('State/UT', fontsize=12)
plt.ylabel('Incidents per 1 Lakh Population', fontsize=12)
plt.title('Accidents, Fatalities, and Injuries on Single-Lane Roads per 1 Lakh Population (Top 15 States/UTs)', fontsize=16)
plt.xticks([i + width for i in x], df_single_lane_sorted['State/UT'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# data of the top 15 states
print("\
Top 15 States/UTs by Total Incidents on Single-Lane Roads per 1 Lakh Population:")
print(df_single_lane_sorted.to_string(index=False))

#correlation between accidents, fatalities, and injuries
correlation = df_single_lane.iloc[:, 1:4].corr()
print("\
Correlation between Accidents, Fatalities, and Injuries on Single-Lane Roads:")
print(correlation)
```

```python
# 27. Number of Accidents happening in DAY and NIGHT TIME for 2014 and 2016.

print(df7.head())


# Calculating data for day and night accidents for 2014
day_2014 = df7['06-900hrs - Day - 2014'].sum() + df7['09-1200hrs - Day - 2014'].sum() + \
           df7['12-1500hrs - Day - 2014'].sum() + df7['15-1800hrs - Day - 2014'].sum()

night_2014 = df7['18-2100hrs - Night - 2014'].sum() + df7['21-2400hrs - Night - 2014'].sum() + \
             df7['00-300hrs - Night - 2014'].sum() + df7['03-600hrs - Night - 2014'].sum()

# Calculate day and night accidents for 2016
day_2016 = df7['06-900hrs - (Day) - 2016'].sum() + df7['09-1200hrs - (Day) - 2016'].sum() + \
           df7['12-1500hrs - (Day) - 2016'].sum() + df7['15-1800hrs - (Day) - 2016'].sum()

night_2016 = df7['18-2100hrs - (Night) - 2016'].sum() + df7['21-2400hrs - (Night) - 2016'].sum() + \
             df7['00-300hrs - (Night) - 2016'].sum() + df7['03-600hrs - (Night) - 2016'].sum()

# Creating a bar plot
data = {
    'Time': ['Day', 'Night', 'Day', 'Night'],
    'Year': ['2014', '2014', '2016', '2016'],
    'Accidents': [day_2014, night_2014, day_2016, night_2016]
}

plt.figure(figsize=(10, 6))
colors = ['lightblue', 'navy', 'lightblue', 'navy']
bars = plt.bar(range(4), data['Accidents'], color=colors)
plt.xticks(range(4), [f"{t}\
{y}" for t, y in zip(data['Time'], data['Year'])])
plt.title('Number of Accidents: Day vs Night (2014 and 2016)')
plt.ylabel('Number of Accidents')

# Adding values labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Printing  the summary
print("\
Summary of Accidents:")
print(f"2014 Day Accidents: {day_2014:,}")
print(f"2014 Night Accidents: {night_2014:,}")
print(f"2016 Day Accidents: {day_2016:,}")
print(f"2016 Night Accidents: {night_2016:,}")
```
